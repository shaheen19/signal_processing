## Automated License Plate Reader Data
The original file as shown in cell 2 is comprised of 21 columns with lot of information on the vehicle type, make, model and color with statistical analysis. 
In this notebook three columns containing time when the picture was captured and license plate information with vehicle type was extracted. 
The simplified version of ALPR file was saved and used for further analysis to identify high emitting vehicles. 

```python
### Import packages
import os
import numpy as np
import pandas as pd
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import seaborn as sns
import datetime as dt
# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
os.getcwd()
```

```python
# Import data file 
df = pd.read_csv('./data/2022-09-02/alpr.csv')
df.head(10)
```

```python
# Check data in the 'best_uuid' columns to extract relevant information.
# print(is_string_dtype(df['best_uuid']))
# print(is_numeric_dtype(df['best_uuid']))
df['Count'] = df['best_uuid'].str.len()
df.head(3)
```

```python
# Extract last 10 character from the column
# Drop data section for plotting and extract timestamps hh:mm:ss
df['license']=df['best_uuid'].str.extract('.*(\d{10})', expand = False)
df['time2'] = df['Time'].astype(str).str.split().str[1]
df.head(3)
```

```python
df1= df[['Time','time2','license', 'vehicle-make-name']]

print(df1.dtypes)
print(df1.shape)
print(df1.info())
df1.head()
```

```python
print('Empty cells/ zero values:', df1.isnull().sum())
```

```python
df1.to_csv('./Results2/simplified_ALPR.csv')
```
